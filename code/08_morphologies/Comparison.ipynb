```python
import numpy as np
import pylab as plt
%matplotlib inline

def plot_comparison(l5,pv,mt,pk,gr):
    cells = [l5, pv, mt, pk, gr]
    cols = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00']
    names = ['L5 pyr.', 'PV cell', 'Martinotti', 'Purkinje','Granule']

    fig = plt.figure()
    for tradeoff,col in zip(cells,cols):
        plt.plot(tradeoff[:,0],tradeoff[:,1],'-',color=col,lw=3)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlim([1e1,1e5])
    plt.ylim([1,150])
    plt.xlabel('mins to reach 5% of ss')
    plt.ylabel('mean percent error')
    plt.legend(names,loc='best')
    return fig
```

# Without Reattachment

```python
l5 = np.loadtxt('./data/no_reattach/l5cell_tradeoff.txt')
pv = np.loadtxt('./data/no_reattach/pvcell_tradeoff.txt')
pk = np.loadtxt('./data/no_reattach/purkinje_tradeoff.txt') 
mt = np.loadtxt('./data/no_reattach/martinotti_tradeoff.txt')
gr = np.loadtxt('./data/no_reattach/granule_tradeoff.txt')

fig = plot_comparison(l5,pv,mt,pk,gr)
plt.title('tradeoff curves without reattachment')
fig.savefig('comparison_no_reattach.eps')
fig.show()
```

# With Reattachment

```python
l5 = np.loadtxt('./data/with_reattach/l5cell_tradeoff.txt')
pv = np.loadtxt('./data/with_reattach/pvcell_tradeoff.txt')
pk = np.loadtxt('./data/with_reattach/purkinje_tradeoff.txt') 
mt = np.loadtxt('./data/with_reattach/martinotti_tradeoff.txt')
gr = np.loadtxt('./data/with_reattach/granule_tradeoff.txt')

fig = plot_comparison(l5,pv,mt,pk,gr)
plt.title('tradeoff curves with reattachment')
fig.savefig('comparison_with_reattach.eps')
fig.show()
```

# Colormaps for animations

```python
from palettable.cubehelix import Cubehelix
from matplotlib.pyplot import cm

u_cmap = cm.cool
us_cmap = Cubehelix.make(start=1, min_light=0.0, max_light=0.9, max_sat=6.0, rotation=0.13).mpl_colormap
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

plt.figure(figsize=(10,2))
ax = plt.subplot(211)
ax.imshow(gradient, aspect='auto', cmap=u_cmap), plt.xticks([]), plt.yticks([])
ax = plt.subplot(212)
ax.imshow(gradient, aspect='auto', cmap=us_cmap), plt.xticks([]), plt.yticks([])
plt.savefig('colbars.eps')
plt.show()
```
