```python
import numpy as np
import pylab as plt
%matplotlib inline

l5 = np.loadtxt('./data/l5cell_tradeoff.txt')
pv = np.loadtxt('./data/pvcell_tradeoff.txt')
pk = np.loadtxt('./data/purkinje_tradeoff.txt') 
mt = np.loadtxt('./data/martinotti_tradeoff.txt')
gr = np.loadtxt('./data/granule_tradeoff.txt')

cells = [l5, pv, mt, pk, gr]
cols = ['#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00'] #['#94ff7b', '#cf2154', '#11b2e8', '#532e37']
names = ['L5 pyr.', 'PV cell', 'Martinotti', 'Purkinje','Granule']

plt.figure()
for tradeoff,col in zip(cells,cols):
    plt.plot(tradeoff[:,0],tradeoff[:,1],'-',color=col,lw=3)
plt.xscale('log')
plt.yscale('log')
plt.xlim([1e1,1e5])
plt.ylim([1,150])
plt.xlabel('mins to reach 5% of ss')
plt.ylabel('mean percent error')
plt.title('tradeoff curves for diff cell types (diff. coeff. = 10)')
plt.legend(names,loc='best')
plt.savefig('comparison.eps')
plt.show()
```
