```python
from __future__ import division
import numpy as np
from scipy.integrate import odeint
import pylab as plt
%matplotlib inline

N = 8 # number of compartments

# trafficking rate constants, produces uniform gradient
a = 10*np.ones(N-1)
b = 10*np.ones(N-1)

# production/degradation
alpha = 0.3  # production rate in first compartment
c = np.linspace(0.01,0.2,N) # detachment
d = np.ones(N)*0.01 # degradation

# Model in state-space form (udot = A*u + b)
# A is a matrix, with blocks [[A1,0];[diag(c),-diag(d)]]
A1_diag = -c*np.ones(N)
A1_diag[:-1] -= a
A1_diag[1:]  -= b
A1 = np.diag(a, -1) + np.diag(A1_diag) + np.diag(b, 1)

A = np.bmat([[A1, np.zeros((N,N))], [np.diag(c), np.diag(-d)]])
A = np.array(A)
#Define differential equation
def udot(u,tt,alph):
    du = np.dot(A,u)     # trafficking and degradation
    du[0] = du[0] + alph # production rate in soma
    return du
```

```python
t = np.linspace(0,1000)
u = odeint(udot, np.zeros(2*N), t, args = (alpha,))

plt.figure(figsize=(6,6))
plt.subplot(211)
plt.plot(t,u[:,:N])
plt.ylabel('u$_i$',fontweight='bold',fontsize=19)
plt.subplot(212)
plt.plot(t,u[:,N:])
plt.ylabel('u$_i*$',fontweight='bold',fontsize=19)
plt.xlabel('time',fontweight='bold',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,10))

## FIRST PANEL, AVERAGED DATA
plt.subplot(2,2,2)
u_ss = []
for alpha in np.random.uniform(0.3,0.7,20):
    u = odeint(udot, np.zeros(2*N), t, args = (alpha,))
    u_ss.append(u[-1,N:])
u_ss = np.array(u_ss)

plt.errorbar(range(N),np.mean(u_ss,axis=0),yerr=np.std(u_ss,axis=0),fmt='o-',color='k')
plt.xlabel('distance from soma\n(compartment index)',fontweight='bold')
plt.ylabel('concentration',fontweight='bold')
plt.title('Model: Averaged Data',fontweight='bold')
plt.xlim([-0.5,7.5])
plt.ylim([0,15])
plt.yticks([0,5,10,15])

## SECOND PANEL, SINGLE CELL DATA
import colorsys
plt.subplot(2,2,3)
alpha_range = [0.3, 0.36, 0.38, 0.43, 0.51, 0.55, 0.58, 0.65]

h = 0    
for (i,alpha) in enumerate(alpha_range):
    u = odeint(udot, np.zeros(2*N), t, args = (alpha,))
    plt.plot(u[-1,N:],'-o',color=colorsys.hsv_to_rgb((h%1),0.8,0.95))
    h += 0.618033988749895 # silver ratio

plt.xlabel('distance from soma\n(compartment index)',fontweight='bold')
plt.ylabel('concentration',fontweight='bold')
plt.title('Model: Single-Cell Data',fontweight='bold')
plt.xlim([-0.5,8.5])
plt.xticks(range(8))
plt.ylim([0,15])
plt.yticks([0,5,10,15])

## THIRD PANEL, EXPERIMENTAL PREDICTIONS (knock-down and over-expression)
plt.subplot(2,2,4)
u_ss = []
for alpha in np.random.uniform(0.6,0.9,50):
    u = odeint(udot, np.zeros(2*N), t, args = (alpha,))
    u_ss.append(u[-1,N:])
u_ss = np.array(u_ss)
plt.errorbar(range(N),np.mean(u_ss,axis=0),yerr=np.std(u_ss,axis=0),fmt='o-',color=(1,0.3,0.3))

u_ss = []
for alpha in np.random.uniform(0.3,0.6,50):
    u = odeint(udot, np.zeros(2*N), t, args = (alpha,))
    u_ss.append(u[-1,N:])
u_ss = np.array(u_ss)
plt.errorbar(range(N),np.mean(u_ss,axis=0),yerr=np.std(u_ss,axis=0),fmt='o-',color='k')

u_ss = []
for alpha in np.random.uniform(0,0.3,50):
    u = odeint(udot, np.zeros(2*N), t, args = (alpha,))
    u_ss.append(u[-1,N:])
u_ss = np.array(u_ss)
plt.errorbar(range(N),np.mean(u_ss,axis=0),yerr=np.std(u_ss,axis=0),fmt='o-',color=(0.3,0.3,1))

plt.xlabel('distance from soma\n(compartment index)',fontweight='bold')
plt.ylabel('concentration',fontweight='bold')
plt.title('Model: Example knock-down and\n over-expression experiments',fontweight='bold')
plt.legend(('over-expression','baseline','knock-down'),loc='upper left')
plt.xlim([-0.5,7.5])
plt.tight_layout()

plt.savefig('magee_model_sushi_alternative.svg')
plt.show()
```
