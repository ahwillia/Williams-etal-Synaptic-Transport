```python
from __future__ import division
import numpy as np

N = 8 # number of compartments

# trafficking rate constants, produces linear gradient
a = np.arange(N-1)+2
b = np.arange(N-1)+1

# Define differential equation
t = np.linspace(0,1000)
def udot(u,tt,alph):
    du = np.dot(A,u)     # trafficking and degradation
    du[0] = du[0] + alph # production rate in soma
    return du
```

```python
from scipy.integrate import odeint
import pylab as plt
%matplotlib inline
plt.figure(figsize=(15,10))

cols = [(0,0,0),(0.5,0.1,0.1),(1,0.2,0.2)]
# Vary the production rate. Show representative single-cell data
alpha_range = [0.1,0.2,0.3]
beta_range = [0.0001,0.001,0.01,0.1,0.2,1.0]
for (s,beta) in enumerate(beta_range):
    # Model in state-space form (udot = A*u + b)
    mid = -beta*np.ones(N)
    mid[:-1] -= a
    mid[1:]  -= b
    A = np.diag(a, -1) + np.diag(mid) + np.diag(b, 1)
    plt.subplot(2,3,s+1)
    for (i,alpha) in enumerate(alpha_range):
        u = odeint(udot, np.zeros(N), t, args = (alpha,))
        plt.plot(u[-1,:],'-o',color=cols[i])
    plt.xlabel('distance from soma\n(compartment index)',fontweight='bold')
    plt.ylabel('$u_i$',fontweight='bold',fontsize=18)
    plt.title(r'$\beta$ = '+str(beta),fontweight='bold',fontsize=18)
    plt.xlim([-0.5,7.5])

legh = plt.legend(alpha_range,title=r'$\alpha$')
plt.setp(legh.get_title(),fontsize='xx-large')
plt.tight_layout()
plt.savefig('vary_beta.png')
plt.show()
```
