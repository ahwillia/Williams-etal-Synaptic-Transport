```python
from __future__ import division
import numpy as np
from scipy.linalg import expm, eig, lstsq
import pylab as plt
%matplotlib inline
```

```python
def run_sim(A,dt=0.5,tmax=None):
    """
    Run a simulation given a matrix A that describes trafficking dynamics in discrete time.
    """
    w,v = eig(A)

    # Find steady-state solution (eigenvector with zero eigenvalue)
    ss_ind = np.argmin(abs(w))
    u_ss = v[:,ss_ind] / sum(v[:,ss_ind]) # re-normalize to sum to 1
    
    # Find slowest mode of system (non-zero eigenvalue with smallest magnitude)
    min_eig = -min(abs(np.delete(w,ss_ind)))
    
    # Initial condition -- molecule starts in soma
    u0 = np.append(1,np.zeros(A.shape[0]-1))
    
    # simulation params
    du = expm(A*dt) # state transition matrix

    # Simulate until all compartments are within 1 percent of steady-state
    t = [0]
    u = [u0]
    
    if tmax is None:
        while max(abs((u[-1]-u_ss)/u_ss)) > 1/(np.e**4):
            u.append(np.dot(du,u[-1]))
            t.append(t[-1]+dt)
    else:
        while t[-1] < tmax:
            u.append(np.dot(du,u[-1]))
            t.append(t[-1]+dt)
    u = np.array(u)

    return (t,u,u_ss,min_eig)
```

```python
## A simple three compartment model to illustrate the bottleneck effect
def bottleneck_model(g=1.0,e=0.01):
    """
    A three compartment model with a bottleneck. The middle compartment has
    large influx rate constants (g) relative to efflux rate constants (e).
    That is, g >> e.
    """
    A = np.array([[-e,      g,      0],
                  [+e ,  -2*g,     +e],
                  [  0,     g,     -e]])
    return A

## Three example traces, showing a bottleneck as e approaches zero
plt.figure(figsize=(5,7))
plt.subplot(311)
t,u,u_ss,min_eig = run_sim(bottleneck_model(e=1.0),dt=0.1,tmax=25)
plt.plot(t,u,'-r')
plt.xlim([t[0],t[-1]])
plt.title('$\epsilon = 1.0$',fontsize=15)

plt.subplot(312)
t,u,u_ss,min_eig = run_sim(bottleneck_model(e=0.5),dt=0.1,tmax=25)
plt.plot(t,u,'-b')
plt.xlim([t[0],t[-1]])
plt.title('$\epsilon = 0.5$',fontsize=15)

plt.subplot(313)
t,u,u_ss,min_eig = run_sim(bottleneck_model(e=0.1),dt=0.1,tmax=25)
plt.plot(t,u,'-g')
plt.xlim([t[0],t[-1]])
plt.title('$\epsilon = 0.1$',fontsize=15)

plt.tight_layout()
plt.show()
```

```python
## General analysis for three compartment example
ee = np.logspace(-2,0,10) 
tmax = []
min_eig = []
for eps in ee:
    t,u,u_ss,minimum_eig = run_sim(bottleneck_model(e=eps),dt=0.1)
    tmax.append(t[-1])
    min_eig.append(minimum_eig)

plt.figure()
plt.plot(ee,-4/(np.array(min_eig)),'-k',label='theory')
plt.plot(ee,tmax,'ok',label='simulation')
plt.legend()
plt.xlabel('$\epsilon/\gamma$',fontsize=23)
plt.ylabel('time to convergence \n (within $1/e^4$ of steady state)',fontsize=12)
plt.show()
```

```python
def rand_model(N,B,kmin=0.1,kmax=1.0):
    """
    Returns a matrix A that defines the transport model (du/dt = A*u)
    
    The model has N chemical compartments and B branch points (randomly positioned)
    
    The trafficking rate constants are drawn from a random uniform distribution on the interval [kmin,kmax)
    """
    assert B <= np.floor((N-2)/2)
    
    # Start with a matrix describing transport in a unbranched cable
    a = np.random.uniform(kmin,kmax,N-1)
    b = np.random.uniform(kmin,kmax,N-1)
    A = construct_tridiag(a,b)
    
    # We will now iteratively add branches by moving elements in the matrix A
    BranchPoints = [] # list of the indices of branch points
    Leaves = [N-1]    # list of the indices of leaves (tips of dendrites)
    
    while len(BranchPoints) < B:
        cL = [x for x in range(2,N-1) if x not in (BranchPoints+Leaves)] # candidate leaves
        i = np.random.choice(cL) # New Leaf
        
        cB = [x for x in range(1,i) if x not in (BranchPoints+Leaves)] # candidate branch points
        if len(cB)==0:
            continue
        j = np.random.choice(cB) # New Branch Point
        
        # Disconnect node i to i+1, node i is now a leaf
        A[i,i] += a[i]    # erase transport from i into i+1
        A[i+1,i] -= a[i]
        A[i,i+1] -= b[i]  # erase transport from i+1 into i
        
        # Connect node j to node i+1
        A[j,j] -= a[i]    # add transport from j into i+1
        A[i+1,j] += a[i]
        A[j,i+1] += b[i]  # add transport from i+1 into j
        
        # Update list of Leaves and Branches
        Leaves.append(i)
        BranchPoints.append(j)
    
    return A

def construct_tridiag(a,b):
    """
    Constructs a tridiagonal matrix describing transport in a cable.
    
    Inputs a and b are lists of trafficking rate constants (N-1 elements each)
    """
    A1 = b
    A2 = -np.append(a,0)-np.append(0,b)
    A3 = a
    return np.diag(A1, 1) + np.diag(A2, 0) + np.diag(A3, -1)
```

```python
t,u,u_ss,min_eig = run_sim(rand_model(10,0))

plt.figure()
plt.plot(t,u)
plt.xlim([t[0],t[-1]])
plt.show()
```

```python
plt.figure()
uss_big = np.tile(u_ss,(u.shape[0],1))
plt.plot((u-uss_big)/uss_big)
plt.xlim([0,100])
plt.show()
```

```python
min_eig_list = []
row_mag = []
diags = []
t_max = []
ss = []

for a in range(1000):
    A = rand_model(10,2)
    t,u,u_ss,min_eig = run_sim(A)
    min_eig_list.append(min_eig)
    row_mag.append(np.sum(abs(A),axis=0))
    diags.append(np.diag(A))
    t_max.append(t[-1])
    ss.append(u_ss)

row_mag = np.array(row_mag)
min_eig_list = np.array(min_eig_list)
ss = np.array(ss)

plt.figure()
plt.plot(abs(min_eig_list),t_max,'ob')
plt.title('The min eigenvalue approximates convergence rate')
plt.show()
```
